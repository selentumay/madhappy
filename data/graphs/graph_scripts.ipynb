```python
import pandas as pd
import numpy as np
import gzip
import matplotlib.pyplot as plt
```

```python
DEFAULT_TRAIN_FILEPATH = ['../train_data_1.gz', '../train_data_2.gz', '../train_data_3.gz', '../train_data_4.gz']
DEFAULT_TEST_FILEPATH = '../test_data.gz'
DEFUALT_VALIDATION_FILEPATH = '../validation_data.gz'
EMOTION_CLASSIFICATION = {0: 'Angry', 1: 'Digust', 2: 'Fear', 3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'}
```

```python
train_data_list = []
for file_path in DEFAULT_TRAIN_FILEPATH:
    temp = pd.read_csv(file_path, compression='gzip')
    train_data_list.append(temp)

train_data = pd.concat(train_data_list)
train_data = train_data[['pixels', 'emotion']]
```

```python
test_data = pd.read_csv(DEFAULT_TEST_FILEPATH, compression='gzip')
test_data = test_data[['pixels', 'emotion']]
```

```python
val_data = pd.read_csv(DEFUALT_VALIDATION_FILEPATH, compression='gzip')
val_data = val_data[['pixels', 'emotion']]
```

```python
train_data['emotion'] = train_data['emotion'].map(EMOTION_CLASSIFICATION)
test_data['emotion'] = test_data['emotion'].map(EMOTION_CLASSIFICATION)
val_data['emotion'] = val_data['emotion'].map(EMOTION_CLASSIFICATION)
```

```python
train_classification_distribution = train_data['emotion'].value_counts().reset_index()
train_classification_distribution = train_classification_distribution.rename(columns={'index': 'emotion', 'emotion': 'train_data'})

test_classification_distribution = test_data['emotion'].value_counts().reset_index()
test_classification_distribution = test_classification_distribution.rename(columns={'index': 'emotion', 'emotion': 'test_data'})

val_classification_distribution = val_data['emotion'].value_counts().reset_index()
val_classification_distribution = val_classification_distribution.rename(columns={'index': 'emotion', 'emotion': 'validation_data'})
```

```python
classification_distribution = train_classification_distribution.merge(val_classification_distribution)
classification_distribution = classification_distribution.merge(test_classification_distribution)
ax = classification_distribution.plot.bar(x='emotion', y=['train_data', 'validation_data', 'test_data'], color=['indigo', 'violet', 'fuchsia'])
ax.set_ylabel('Number of Images')
ax.set_xlabel('Emotion')
ax.set_title('FER2013: Number of Data Points Per Emotion')
ax.legend()
ax.figure.savefig('data_classification_distribution.png')
```

```python
def get_image(row):
    pixels, emotion = row['pixels'], row['emotion']
    img = np.array(pixels.split())
    img = img.reshape(48,48)
    image = np.zeros((48,48,3))
    image[:,:,0] = img
    image[:,:,1] = img
    image[:,:,2] = img
    return image.astype(np.uint8), emotion
```

```python
plt.figure(0, figsize=(16,10))
plt.title('FER2013: Sample of Images')
for i in range(1,10):
    img, emotion = get_image(train_data.iloc[i])
    plt.subplot(3,3,i)
    plt.title(emotion)
    plt.imshow(img)
plt.tight_layout()
plt.savefig('image_sample.png')
```
